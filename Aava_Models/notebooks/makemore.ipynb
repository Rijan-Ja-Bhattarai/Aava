```python
from pathlib import Path 
import requests
```

```python
path = Path("../data/names.txt")

response = requests.get("https://raw.githubusercontent.com/karpathy/makemore/refs/heads/master/names.txt")

with open(path, "w") as f:
    f.write(response.text)
print("Complete")
```

```python
word = open(path).read().splitlines()
word[:10]
```

```python
len(word), min(len(w) for w in word)
```

```python
for w in word[:1]:
    for ch1, ch2 in zip(w, w[1:]):
        print(ch1, ch2)
```
